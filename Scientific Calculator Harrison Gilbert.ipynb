## Scientific Calculator

```python
import math
class SciCalculator():
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        
    def add(number1, number2):
        return(number1 + number2)
        
    def sub(number1, number2):
        return(number1 - number2)
        
    def mult(number1, number2):
        return(number1*number2)
    
    def div(number1, number2):
        return(number1/number2)
        
    def sqrt(num):
        return(math.sqrt(num))
        
    def power(root, exp):
        return(root**exp)
        
    def remainder(num1, num2):
        return( math.remainder(num1, num2))
        
    def log(x, base):
        return(math.log(x,base))
        
    def natLog(x):
        return(math.log1p(x))
        
    def asin(x):
        return(math.asin(x))
    
    def acos(x):
        return(math.acos(x))
        
    def atan(x):
        return(math.atan(x))
        
    def sin(x):
        return(math.sin(x))
    
    def cos(x):
        return(math.cos(x))
        
    def tan(x):
        return(math.tan(x))
    
    def degrees(x):
        return(math.degrees(x))
    
    def radians(x):
        return(math.radians(x))
        
    def pi():
        return(str(math.pi))
    
    def e():
        return(math.e)
    
    
while True:
    print("\nSelect a number: ")
    print("1: Addition")
    print("2: Subtraction")
    print("3: Multiplication")
    print("4: Divisiion")
    print("5: Squareroot")
    print("6: Power")
    print("7: Remainder")
    print("8: Log")
    print("9: Natural Log")
    print("10: Arcsin")
    print("11: Arccos")
    print("12: Arctan")
    print("13: Sin")
    print("14: Cos")
    print("15: Tan")
    print("16: Degrees")
    print("17: Radians")
    print("18: Pi")
    print("19: E")
    print("20: Exit")
    choice = int(input("Enter your choice (1-20): "))
    
    if choice == 1:
        num1 = int(input("Enter number 1: "))
        num2 = int(input("Enter number 2: "))
        print("Addition: ", SciCalculator.add(num1, num2))
    elif choice == 2:
        num1 = int(input("Enter number 1: "))
        num2 = int(input("Enter number 2: "))
        print("Subtraction: ", SciCalculator.sub(num1, num2))
    elif choice == 3:
        num1 = int(input("Enter number 1: "))
        num2 = int(input("Enter number 2: "))
        print("Multiplication: ", SciCalculator.mult(num1, num2))
    elif choice == 4:
        num1 = int(input("Enter number 1: "))
        num2 = int(input("Enter number 2: "))
        print("Division: ", SciCalculator.div(num1, num2))
    elif choice == 5:
        num1 = int(input("Enter number: "))
        print("Squareroot: ", SciCalculator.sqrt(num1))
    elif choice == 6:
        num1 = int(input("Enter base: "))
        num2 = int(input("Enter exponent: "))
        print("Subtraction: ", SciCalculator.power(num1, num2))
    elif choice == 7:
        num1 = int(input("Enter number 1: "))
        num2 = int(input("Enter number 2: "))
        print("Remainder: ", SciCalculator.remainder(num1, num2))
    elif choice == 8:
        num1 = int(input("Enter number 1: "))
        num2 = int(input("Enter number 2: "))
        print("Log: ", SciCalculator.log(num1, num2))
    elif choice == 9:
        num1 = int(input("Enter number 1: "))
        print("NaturalLog: ", SciCalculator.natLog(num1))
    elif choice == 10:
        num1 = int(input("Enter number: "))
        print("Arcsin: ", SciCalculator.asin(num1))
    elif choice == 11:
        num1 = int(input("Enter number: "))
        print("Arccos: ", SciCalculator.acos(num1))
    elif choice == 12:
        num1 = int(input("Enter number: "))
        print("Arctan: ", SciCalculator.atan(num1))
    elif choice == 13:
        num1 = int(input("Enter number: "))
        print("Sin: ", SciCalculator.sin(num1))
    elif choice == 14:
        num1 = int(input("Enter number: "))
        print("Cos: ", SciCalculator.cos(num1))
    elif choice == 15:
        num1 = int(input("Enter number: "))
        print("Tan: ", SciCalculator.tan(num1))
    elif choice == 16:
        num1 = int(input("Enter number: "))
        print("Degrees: ", SciCalculator.degrees(num1))
    elif choice == 17:
        num1 = int(input("Enter number: "))
        print("Radians: ", SciCalculator.radians(num1))
    elif choice == 18:
        print("Pi: ", SciCalculator.pi())
    elif choice == 19:
        print("E: ", SciCalculator.e())
    elif choice == 20:
        break
    else:
        print("Invalid Number")
        
```
